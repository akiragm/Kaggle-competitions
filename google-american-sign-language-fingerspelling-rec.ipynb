##やりたいこと
このコンペティションの目標は、アメリカンサインランゲージ（ASL）の指文字を検出し、テキストに変換することです。<br>
したがって、ASLの指文字を検出および翻訳するためのモデルを構築する必要があります。<br>
<br>
与えられたデータには、以下のようなファイルが含まれています。<br>
<br>
[train/supplemental_metadata].csv:メタデータが含まれているCSVファイルです。<br>
このファイルには、ランドマークファイルへのパス、データファイルの一意の識別子（file_id）、データ貢献者の一意の識別子（participant_id）、<br>
ランドマークシーケンスの一意の識別子（sequence_id）、およびランドマークシーケンスのラベル（phrase）が含まれています。<br>
<br>
character_to_prediction_index.json:文字と予測インデックスの対応が記載されているJSONファイルです。<br>
<br>
[train/supplemental]_landmarks/:ランドマークデータが含まれているディレクトリです。<br>
ここには、MediaPipeのホリスティックモデルを使用して抽出されたランドマークが含まれています。<br>
各データファイルには複数のシーケンスが含まれる場合があります。各シーケンスはフレームごとにランドマークの座標が含まれるため、テキストに変換するための情報を利用できます。<br><br>

これらの情報を考慮して、ASLの指文字を検出および翻訳するためのモデルを構築する方法について説明します。<br><br>

##1<br>
ランドマークデータを準備する:[train/supplemental]_landmarks/ディレクトリ内のファイルから、各フレームのランドマーク座標データを読み込みます。<br>
このデータは、MediaPipeホリスティックモデルによって抽出されたものです。各ランドマークにはx、y、z座標が含まれており、手の形状や位置を表しています。<br>
##2<br>
ラベルデータを準備する:[train/supplemental_metadata].csvファイルから、各シーケンスのラベルデータ（指文字のテキスト）を読み込みます。<br>
これは、ランドマークデータに対応する正解ラベルとなります。<br>
##3<br>
モデルのトレーニング:TensorFlowや他のフレームワークを使用して、ASLの指文字を検出および翻訳するためのモデルをトレーニングします。<br>
入力としては、各フレームのランドマーク座標データを使用し、出力としては各フレームの指文字のテキストを予測するモデルを構築します。<br>
トレーニングには、適切なネットワークアーキテクチャ（たとえば、畳み込みニューラルネットワークやリカレントニューラルネットワーク）や損失関数、最適化アルゴリズムなどを選択します。<br>
また、ランドマークデータと対応する正解ラベルを使用してモデルをトレーニングします。<br>
##4<br>
モデルの変換:トレーニングされたモデルをTensorFlowLiteモデルに変換します。<br>
これにより、モデルを軽量化し、モバイルデバイスやエッジデバイスで効率的に実行できるようになります。<br>
変換には、TensorFlowLiteの変換ツールやAPIを使用します。<br>
##5<br>
提出準備:変換されたTensorFlowLiteモデルを提出するために準備します。<br>
提出には、コンペティションの要件に従って、モデルのチェックポイントをtflite形式に変換する必要があります。<br>
詳細な手順は評価ページで提供されるはずです。<br>
##6<br>
モデルの評価とチューニング:提出したモデルは評価され、性能が評価されるでしょう。<br>
コンペティションの評価基準に従って、モデルの精度や効率性が評価されます。<br>
もし必要ならば、モデルを改善するためにトレーニングやハイパーパラメータの調整を行うことができます。<br><br>

```python
#1 各単語を用意し、画像からランドマークを抽出　
import os
import cv2
import mediapipe as mp

mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

def get_landmark_coordinates(folder_path):
    file_list = os.listdir(folder_path)
    image_files = [file for file in file_list if file.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    with mp_holistic.Holistic(static_image_mode=True) as holistic:
        for idx, file in enumerate(image_files):
            image_path = os.path.join(folder_path, file)
            image = cv2.imread(image_path)
            image_height, image_width, _ = image.shape
            # 処理する前にBGR画像をRGBに変換
            results = holistic.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

            if results.pose_landmarks:
                print(
                    f'Nose coordinates in {file}: ('
                    f'{results.pose_landmarks.landmark[mp_holistic.PoseLandmark.NOSE].x * image_width}, '
                    f'{results.pose_landmarks.landmark[mp_holistic.PoseLandmark.NOSE].y * image_height})'
                )
            # 画像に姿勢と左手と右手と顔のランドマークを描画
            annotated_image = image.copy()
#             mp_drawing.draw_landmarks(
#                 annotated_image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION)
            mp_drawing.draw_landmarks(
                annotated_image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
            mp_drawing.draw_landmarks(
                annotated_image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
            # Use mp_holistic.UPPER_BODY_POSE_CONNECTIONS for drawing below when
            # upper_body_only is set to True.
#            mp_drawing.draw_landmarks(
#                annotated_image, results.pose_landmarks, mp_holistic.UPPER_BODY_POSE_CONNECTIONS)
            
            annotated_image_path = os.path.join(folder_path, 'annotated_image' + str(idx) + '.png')
            cv2.imwrite(annotated_image_path, annotated_image)

# 使用例
folder_path = "/kaggle/input/aslpicture"
get_landmark_coordinates(folder_path)
```

```python
##keras
#sequence_idとphraseが対応し、ラベルがphrase
#126frame
#x,Y,Z


importtensorflowastf
fromtensorflow.keras.modelsimportSequential
fromtensorflow.keras.layersimportDense

defcreate_model(input_size,hidden_size,num_classes):
#モデルの定義
model=Sequential()
model.add(Dense(hidden_size,activation='relu',input_shape=(input_size,)))

#隠れ層の追加
model.add(Dense(hidden_size,activation='relu'))
model.add(Dense(hidden_size,activation='relu'))
model.add(Dense(hidden_size,activation='relu'))

#出力層の追加
model.add(Dense(num_classes,activation='softmax'))

returnmodel
```

```python
#LightBGM
importlightgbmaslgb
fromsklearn.datasetsimportload_breast_cancer
fromsklearn.model_selectionimporttrain_test_split
fromsklearn.metricsimportaccuracy_score

#データの読み込み
data=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.2,random_state=42)

#LightGBMのモデルを定義
model=lgb.LGBMClassifier()

#モデルのトレーニング
model.fit(X_train,y_train)

#テストデータで予測を実行
y_pred=model.predict(X_test)

#精度の評価
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)

```
